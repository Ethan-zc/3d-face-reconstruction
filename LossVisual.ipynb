```python
file1 = open('loss_log.txt', 'r')
Lines = file1.readlines()
```

```python
count = 0
# Strips the newline character
all_loss = []
img_loss = []
lm_loss = []

for line in Lines:
    count += 1
    all_start = int(line.find("all"))
    all_start_end = int(all_start + 3)
    color = int(line.find("color"))
    color_end = int(color+5)
    lm = int(line.find("lm"))
    lm_end = int(lm+2)
    
    if all_start != -1:
        print("Line{}: {}".format(count, line.strip()))
        all_loss_single = float(line[all_start_end+2:all_start_end+7])
        all_loss.append(all_loss_single)
        
        img_loss_single = float(line[color_end+2:color_end+7])
        img_loss.append(img_loss_single)
        
        lm_loss_single = float(line[color_end+2:color_end+7])
        lm_loss.append(lm_loss_single)
# print(all_loss)
```

```python
print(lm_loss)
```

```python
import matplotlib.pyplot as plt
plt.plot(all_loss)
plt.ylabel("Total loss")
```

```python
plt.plot(img_loss)
plt.ylabel("Image loss")
```
